doing this ways is called pyplot API, but pyplot API is generally less-flexible than the object-oriented API

### Object-oriented API

```python
# 1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # add some data
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
print(type(fig)); print(type(ax))
```

```python
fig, ax = plt.subplots(figsize=(10, 10)) # width, height
ax.plot(x, y)
ax.set(title='Simple plot', xlabel='x-axis', ylabel='y-axis')
plt.show()
fig.savefig('simple_plot.png')
```

# Making figures with numpy array
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
#Create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
# Creating a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# Creating a bar plot by dictionary
butter_prices = {'Almond butter': 12,
                 'Peanut butter': 10,
                 'Cashew butter': 9}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title='Butter prices', ylabel='Price ($)')
plt.show()
```

```python
# Horizontal bar
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
plt.show()
```

```python
# Creating histogram
x = np.random.randn(1000) #samples from a normal distribution
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

### Two options for subplot

```python
# 1st option
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(butter_prices.keys(), butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# 2nd option
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(butter_prices.keys(), butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

Calling directly plot function is same as using pyplot API. Now, we will shift into OO API method with Pandas DataFrame

### OO Method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
```

```python
len(over_50)
```

```python
#Pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target')
plt.show()
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
ax.set_xlim([45, 100]) # x range 45 - 100
plt.show()
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
ax.set(title='Heart Disease and Cholestorel Level', xlabel='Age', ylabel='Cholestoral')
ax.legend(*scatter.legend_elements(), title='Target')
ax.axhline(over_50['chol'].mean(), linestyle='--')
plt.show()
```

```python
over_50.target.values
```

```python
#Subplot of age, chol, and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True) #two plot have the same x value

scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
ax0.set(title='Heart Disease and Cholestoral Level', ylabel='Cholestoral')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])
ax1.set(title='Heart Disease and Thalach Level', xlabel='Age', ylabel='Thalach')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

#Set fig title
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
plt.show()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True) #two plot have the same x value

scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='winter')
ax0.set(title='Heart Disease and Cholestoral Level', ylabel='Cholestoral')
ax0.set_xlim([50, 80])
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='winter')
ax1.set(title='Heart Disease and Thalach Level', xlabel='Age', ylabel='Thalach')
ax.set_ylim([60, 200])
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

#Set fig title
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
plt.show()
```
